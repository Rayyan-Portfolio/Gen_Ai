```python
import gradio as gr
from transformers import pipeline

# Load Whisper model
pipe = pipeline("automatic-speech-recognition", model="openai/whisper-large-v3-turbo")

# Function to process audio and return transcription
def transcribe(audio):
    result = pipe(audio, return_timestamps=True, generate_kwargs={"language": ["English"]})
    return result["text"]  # Extract the transcribed text

# Gradio UI
app = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Whisper AI Transcription",
    description="Upload an audio file and get transcription."
)

# Run the app
app.launch(pwa=True)
```

```python
import gradio as gr
from transformers import pipeline

# Load Whisper model for speech-to-text
asr_pipeline = pipeline("automatic-speech-recognition", model="openai/whisper-large-v3")

# Load Summarization model
summarization_pipeline = pipeline("summarization", model="facebook/bart-large-cnn")

# Function to transcribe speech to text
def transcribe_and_summarize(audio):
    # Speech-to-text
    transcription_result = asr_pipeline(audio)
    text = transcription_result["text"]

    # Generate summary (only if text is long enough)
    if len(text.split()) > 20:  # Summarization works best for longer texts
        summary_result = summarization_pipeline(text, max_length=50, min_length=20, do_sample=False)
        summary = summary_result[0]["summary_text"]
    else:
        summary = "Text is too short for summarization."

    return text, summary

# Gradio UI
app = gr.Interface(
    fn=transcribe_and_summarize,
    inputs=gr.Audio(type="filepath"),
    outputs=["text", "text"],  # One for transcription, one for summary
    title="Whisper AI Speech-to-Text & Summarization",
    description="Upload an audio file to get both full transcription and a concise summary.",
)

# Run the app with PWA support
app.launch(pwa=True)
```
