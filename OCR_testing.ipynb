```python
# Install EasyOCR (run this cell once)
!pip install --quiet easyocr

# Import necessary libraries
import cv2
import easyocr
import matplotlib.pyplot as plt

# Update the image path to point to your dataset image file on Kaggle
image_path = '/content/2.png'  # Change accordingly

# Read the image using OpenCV
image = cv2.imread(image_path)

# (Optional) Display the image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Input Image")
plt.show()

# Initialize the EasyOCR Reader.
# Specify the languages you need (e.g., ['en'] for English).
reader = easyocr.Reader(['en'])

# Extract text from the image.
# Setting detail=0 returns only the text strings.
results = reader.readtext(image, detail=0)

# Join the detected text strings into one output string.
extracted_text = ' '.join(results)

# Print the extracted text
print("Extracted Text:")
print(extracted_text)
```
